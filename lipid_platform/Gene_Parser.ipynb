```python

# Define empty lists to hold the numbers and gene codes
numbers_for_gene = []
genes = []

# Open the file and parse each line
with open('./Gene_Database/genes.txt', 'r') as file:
    for line in file:
        # Use a regular expression to find the number and gene code
        match = re.search(r'VALUES \((\d+), \'(\w+)\'', line)
        if match:
            # If a match was found, add the number and gene code to the lists
            numbers_for_gene.append(int(match.group(1)))
            genes.append(match.group(2))

# Create a dataframe from the lists
df_genes = pd.DataFrame({
    'Number': numbers,
    'Gene': genes
})

print(df_genes)
```

```python

# Define empty lists to hold the numbers and gene codes
reactions = []
Gene_Code = []

# Open the file and parse each line
# Open the file and parse each line
with open('./Gene_Database/genes2lipid.txt', 'r') as file:
    for line in file:
        # Use a regular expression to find the reaction number and gene code
        print(line)
        match = re.search(r'VALUES \((\d+), *(\d+)\)', line)
        if match:
            # If a match was found, add the reaction number and gene code to the lists
            reactions.append(int(match.group(1)))
            Gene_Code.append(int(match.group(2)))

# Create a dataframe from the lists
df_gene_2_reaction = pd.DataFrame({
    'Reaction': reactions,
    'Gene Code': Gene_Code
})

print(df_gene_2_reaction)

print(len(df_gene_2_reaction))
```

```python

# Read the file
with open('./Gene_Database/Lipid Reactions.txt', 'r') as file:    # replace 'filename.txt' with your file name
    text = file.read()

# Regular expression pattern to capture required data
# I've included \s* to match any number of spaces
pattern = r"INSERT INTO biopan_reaction VALUES \((\d+),\s*(.*?),\s*'(.*?)',\s*'(.*?)',"

# Find all matches and create a list of dictionaries
matches = re.findall(pattern, text)
data = [{'reaction_code': int(m[0]), 'Lipid 1': m[1].split(',')[0].replace("E'", ""), 'Lipid 2': m[1].split(',')[1].replace("'", ""), 'Class': m[2], 'reaction_type': m[3]} for m in matches]

# Create pandas DataFrame
df = pd.DataFrame(data)
```
