```python
import pandas as pd
liver = pd.read_csv('Liver_OzON_Truth_Results_16Jan2024.csv')
brainWT = pd.read_csv('BrainWT_OzON_Truth_Results_14Jan2024.csv')
brain5xFAD = pd.read_csv('Brain5xFAD_OzON_Truth_Results_14Jan2024.csv')

liver
brainWT
brain5xFAD
```

drop columns to clean up dataframe for caitlin

```python
columns_to_drop = ['SLBG_group', 'Cluster_Ground_Truth_Dif', 'Cluster_RT', 'Group', 'Match_Group','Group_Sample']

# Dropping the specified columns from each DataFrame
liver = liver.drop(columns=columns_to_drop, errors='ignore')
brainWT = brainWT.drop(columns=columns_to_drop, errors='ignore')
brain5xFAD = brain5xFAD.drop(columns=columns_to_drop, errors='ignore')

brain5xFAD
```

```python
# Original DataFrames: liver, brainWT, brain5xFAD
# Creating new DataFrames with '_summary' appended to each variable name
liver_summary = liver.copy()
brainWT_summary = brainWT.copy()
brain5xFAD_summary = brain5xFAD.copy()

# Filtering out rows where 'db' is 'n-7' in each of the summary dataframes
liver_summary = liver_summary[liver_summary['db'] != 'n-7']
brainWT_summary = brainWT_summary[brainWT_summary['db'] != 'n-7']
brain5xFAD_summary = brain5xFAD_summary[brain5xFAD_summary['db'] != 'n-7']

# Dropping the 'OzESI_Intensity' column from each of the summary dataframes
liver_summary = liver_summary.drop(columns=['OzESI_Intensity','db'])
brainWT_summary = brainWT_summary.drop(columns=['OzESI_Intensity','db'])
brain5xFAD_summary = brain5xFAD_summary.drop(columns=['OzESI_Intensity','db'])

# Define the new order for the columns
new_column_order = ['Lipid', 'Area_DDT_Ratio'] + [col for col in liver_summary.columns if col not in ['Lipid', 'Area_DDT_Ratio']]

# Reorder the columns for each DataFrame
liver_summary = liver_summary[new_column_order]
brainWT_summary = brainWT_summary[new_column_order]
brain5xFAD_summary = brain5xFAD_summary[new_column_order]

# Dropping rows with NaN in 'Area_DDT_Ratio' from each of the summary dataframes
liver_summary = liver_summary.dropna(subset=['Area_DDT_Ratio'])
brainWT_summary = brainWT_summary.dropna(subset=['Area_DDT_Ratio'])
brain5xFAD_summary = brain5xFAD_summary.dropna(subset=['Area_DDT_Ratio'])

#save to each dataframe to excel
liver_summary.to_excel('FaceFats_OzESI_Liver_n9n7Ratio_summary.xlsx')
brainWT_summary.to_excel('FaceFats_OzESI_BrainWT_n9n7Ratio_summary.xlsx')
brain5xFAD_summary.to_excel('FaceFats_OzESI_Brain5xFAD_n9n7Ratio_summary.xlsx')



brain5xFAD_summary
```
