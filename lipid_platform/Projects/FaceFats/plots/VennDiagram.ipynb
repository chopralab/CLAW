# Make liver vs brainWT vs Brain5xFAD dataframe column for Venn diagram

```python
def liver_brain_venn(df):
    """
    Function to create a new column 'Venn' in the dataframe based on the 'Biology' and 'Genotype' columns.
    
    Parameters:
    df (DataFrame): The DataFrame to modify.
    
    Returns:
    DataFrame: The DataFrame with the new 'Venn' column.
    """
    # Create a new column 'Venn' based on conditions
    df['Venn'] = df.apply(lambda x: 'Liver' if x['Biology'] == 'liver' else 'Brain' + x['Genotype'], axis=1)
    
    # Return the modified DataFrame
    return df

# Example usage:
# Assuming df is your DataFrame already loaded with the data
df = liver_brain_venn(df)
df

```

```python
import pandas as pd

# Assuming 'df' is your DataFrame and it has the columns 'Lipid' and 'Sample_ID'

# First, we'll create a function to pivot the table and indicate presence or absence of lipids.
def create_pivot_table(df_liver, lipid_col='Lipid', sample_col='Venn'):
    # Create a DataFrame with unique combinations of Lipid and Sample_ID
    unique_combinations = df_liver[[lipid_col, sample_col]].drop_duplicates()
    
    # Add a column 'Presence' to indicate presence of a lipid (will be all 1s)
    unique_combinations['Presence'] = 1

    # Pivot the table to get Sample_IDs as columns and Lipids as rows
    pivot_table = unique_combinations.pivot(index=lipid_col, columns=sample_col, values='Presence')

    # Replace NaN with 0 to indicate absence
    pivot_table = pivot_table.fillna(0).astype(int)

    return pivot_table


# First, we'll create a function to pivot the table and indicate presence or absence of lipids.
def create_pivot_table_bio(df_liver, lipid_col='Lipid', sample_col='Biology'):
    # Create a DataFrame with unique combinations of Lipid and Sample_ID
    unique_combinations = df_liver[[lipid_col, sample_col]].drop_duplicates()
    
    # Add a column 'Presence' to indicate presence of a lipid (will be all 1s)
    unique_combinations['Presence'] = 1

    # Pivot the table to get Sample_IDs as columns and Lipids as rows
    pivot_table = unique_combinations.pivot(index=lipid_col, columns=sample_col, values='Presence')

    # Replace NaN with 0 to indicate absence
    pivot_table = pivot_table.fillna(0).astype(int)

    return pivot_table


pivot_df_liver = create_pivot_table_bio(df_liver)
pivot_df_brain5xFAD = create_pivot_table_bio(df_5xfad)
pivot_df_brainWT = create_pivot_table_bio(df_wt)
pivot_df = create_pivot_table(df)








# #save to csv
# pivot_df_liver.to_csv('Venn/venn_liver.csv')
pivot_df_brain5xFAD.to_csv('Venn/venn_brain5xFAD.csv')
pivot_df_brain5xFAD.to_excel('Venn/venn_brain5xFAD.xlsx')

pivot_df_brainWT.to_csv('Venn/venn_brainWT.csv')
pivot_df_brainWT.to_excel('Venn/venn_brainWT.xlsx')


pivot_df.to_csv('Venn/venn_Liver_BrainWT_Brain5xFAD.csv')
pivot_df.to_excel('Venn/venn_Liver_BrainWT_Brain5xFAD.xlsx')


pivot_df_liver
pivot_df_brain5xFAD
# pivot_df_brainWT
# pivot_df

```
