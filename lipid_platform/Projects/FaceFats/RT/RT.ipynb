```python
import pandas as pd

df = pd.read_excel('facefats_RTs.xlsx')
df = df.sort_values('Lipid')
df
```

```python
# Function to reformat the Lipid column
def reformat_lipid(lipid):
    # Split at the '_'
    parts = lipid.split('_')
    tg_part = parts[0]
    fa_part = parts[1] if len(parts) > 1 else ""

    # Replace '|' with ',' and remove extra brackets
    tg_part = tg_part.replace('|', ',').replace(']', '').replace('[', '')

    return f"[{tg_part}]_{fa_part}"

# Apply the reformatting function to the Lipid column
df['Lipid'] = df['Lipid'].apply(reformat_lipid)


# Function to remove the last parenthesis in the Lipid column
def remove_last_parenthesis(lipid):
    return lipid.replace("))", ")")

# Apply the function to the Lipid column
df['Lipid'] = df['Lipid'].apply(remove_last_parenthesis)

#redo the index
df = df.reset_index(drop=True)
df
```

```python
df.to_csv('FF_Cait_RTs.csv', index=False)
```

```python
lipids = [
    "[TG(42:2)]_FA18:1 | DG(44:9),DG(43:2)_C18:1",
    "[TG(45:3)]_FA16:1",
    "[TG(45:4)]_FA16:1",
    "[TG(50:3)]_FA18:1",
    "[TG(51:6)]_FA16:1",
    "[TG(52:5)]_FA16:1",
    "[TG(52:6)]_FA18:2",
    "[TG(53:6)]_FA18:1",
    "[TG(54:5)]_FA18:1",
    "[TG(54:6)]_FA18:1",
    "[TG(55:10),TG(54:3)]_FA18:1",
    "[TG(55:11),TG(54:4)]_FA16:1",
    "[TG(55:11),TG(54:4)]_FA18:1",
    "[TG(55:8),TG(54:1)]_FA18:1",
    "[TG(55:9),TG(54:2)]_FA18:1",
    "[TG(56:10),TG(55:3)]_FA18:1",
    "[TG(56:11),TG(55:4)]_FA18:1",
    "[TG(56:6)]_FA18:1",
    "[TG(56:7),TG(55:0)]_FA16:1",
    "[TG(56:7),TG(55:0)]_FA18:1",
    "[TG(57:10),TG(56:3)]_FA18:1",
    "[TG(57:11),TG(56:4)]_FA18:1",
    "[TG(57:12),TG(56:5)]_FA18:1",
    "[TG(57:9),TG(56:2)]_FA18:1",
    "[TG(58:7),TG(57:0)]_FA18:1",
    "[TG(59:11),TG(58:4)]_FA18:1",
    "[TG(59:12),TG(58:5)]_FA18:1",
    "[TG(59:13),TG(58:6)]_FA18:1"
]

def match_lipids(row, lipids_list):
    for lipid in lipids_list:
        # Split the lipid string into individual components
        components = lipid.split('|')
        # Check if any component matches the row's lipid
        if any(row['Lipid'] in comp for comp in components):
            return True
    return False

# Apply the function to each row in the DataFrame
df['Match'] = df.apply(match_lipids, args=(lipids,), axis=1)

# Filter to show only the matching rows
matching_rows = df[df['Match'] == True]

matching_rows
```

```python
matching_rows.to_csv('FF_Cait_RTs_matching.csv', index=False)
```
