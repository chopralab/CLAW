```python
import pandas as pd

# List of file names
file_names = [f'FF_OzMatched_{i}.csv' for i in range(1, 11)]

# Initialize an empty DataFrame
combined_df = pd.DataFrame()

# Loop through each file and append its contents to the combined DataFrame
for file in file_names:
    temp_df = pd.read_csv(file)
    combined_df = combined_df.append(temp_df, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('FF_OzMatched_full.csv', index=False)

# Display the combined DataFrame (optional)
combined_df.head(None)
```
