```python
import pandas as pd
```

```python
df = pd.read_csv('FF_RT178_188.csv')
# Filter for unique pairs of Sample_ID, Lipid, and db_pos
max_df = df.loc[df.groupby(['Sample_ID', 'Lipid', 'db_pos'])['OzESI_Intensity'].idxmax()]
unique_df = max_df.drop_duplicates(subset=['Sample_ID', 'Lipid', 'db_pos'])

# Display the filtered DataFrame
unique_df.head(None)

# Define the substrings for the lipids you want to keep
list1 = ['52:6']
list2 = ['51:6', '52:5', '56:7', '54:6']
l13 = ['50:3', '54:5', '56:6', '56:7', '58:7']
l14 = ['55:1', '57:1', '59:1']
l15 = ['52:7', '55:9', '56:1', '57:1', '59:1', '55:1']
l16 = ['55:8', '56:1', '57:1', '59:1']
l17 = ['57:9']

lipids_to_keep = l17

# Create a mask for rows where the 'Lipid' column contains any of the substrings
mask = unique_df['Lipid'].str.contains('|'.join(lipids_to_keep))
filtered_df = unique_df[mask]

db_pos_to_keep = ['n-9', 'n-7', 'n-10', 'n-12']
# Create a mask for rows where the 'db_pos' column contains any of the substrings
mask2 = filtered_df['db_pos'].isin(db_pos_to_keep)

filtered_df2 = filtered_df[mask2]

columns_to_drop = [
    'Unnamed: 0', 'Unnamed: 0.1', 'Class', 'Parent_Ion', 
    'Product_Ion', 'Transition', 'n-7', 'n-9', 'n-10', 'n-12', 'Unnamed: 0.1.1'
]
filtered_df2 = filtered_df2.drop(columns=columns_to_drop)
filtered_df2.to_excel('../filtered_df/FF_filtered_RT178_188.xlsx')

filtered_df2.head(None)
```

```python
import pandas as pd
import os

# Define the directory containing the Excel files
directory_path = '../filtered_df/'

# Initialize an empty DataFrame to hold all the appended data
appended_df = pd.DataFrame()

# Loop through all files in the directory
for filename in os.listdir(directory_path):
    # Check if the file is an Excel file
    if filename.endswith('.xlsx'):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)
        # Read the Excel file into a DataFrame
        df = pd.read_excel(file_path)
        # Append the DataFrame to the aggregated DataFrame
        appended_df = appended_df.append(df, ignore_index=True)

# Now 'appended_df' contains all the data from the Excel files

appended_df.head(None)
```

```python
appended_df2 = appended_df.iloc[:,1:]
appended_df2.head(None)
```

```python

# Lists to search for
biologies = ['cereb', 'cortex', 'hippo', 'dienc', 'liver']
genes = ['5xFAD', 'WT']

# Function to search for substrings in Sample_ID
def find_match(text, search_list):
    for item in search_list:
        if item in text:
            return item
    return None


# Apply the function to create new columns in appended_df2
appended_df2['Biology'] = appended_df2['Sample_ID'].apply(lambda x: find_match(x, biologies))
appended_df2['Gene'] = appended_df2['Sample_ID'].apply(lambda x: find_match(x, genes))


# Display the DataFrame
appended_df2.head(None)
```

```python
sorted_df = appended_df2.sort_values(by=['Lipid', 'Sample_ID', 'Biology', 'Gene','db_pos'])
sorted_df.head(None)
```
