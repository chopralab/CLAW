```python
import pandas as pd
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
#use pandas to import csv file Caitlin_data.csv
df = pd.read_csv('CLAW_paper.csv')
```

```python
df.head(None)
```

```python
# Extract sample names and brain regions from the column names
samples = [col.split('-')[0] + '-' + col.split('-')[1] for col in df.columns if '10x' in col]
samples = list(set(samples))  # To get unique samples

brain_regions = ['hippo', 'diencephalon', 'cortex', 'cerebellum']

# Create an empty dataframe to store results
result_df = pd.DataFrame(columns=['Sample', 'Brain_Region', 'Sum_Intensity'])

for sample in samples:
    for region in brain_regions:
        # Get columns related to the current sample and brain region
        cols = [col for col in df.columns if sample in col and region in col]
        
        # Sum up the intensities for current group
        total_intensity = df[cols].sum().sum()
        
        result_df = result_df.append({
            'Sample': sample,
            'Brain_Region': region,
            'Sum_Intensity': total_intensity
        }, ignore_index=True)

print(result_df)
```

```python
#save as csv file
result_df.to_csv('MRM_results/summed_intensity.csv', index=False)
```
