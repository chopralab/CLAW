```python
import pandas as pd
```

# Max Intensity vs Manual Max Intensity

```python
df1 = pd.read_csv('data/CLAW_Canola_OzON_Ratio_14Dec2023_6pm.csv')
df2 = pd.read_csv('data/CLAW_Canola_OzON_Ratio_524.csv')
#concatenate the two dataframes
df = pd.concat([df1, df2])
df
```

```python
#sort df by Purity and then Lipid
df.sort_values(by=['Purity', 'Lipid'], inplace=True)
#drop Ratio NaNs
df.dropna(subset=['Ratio'], inplace=True)
#drop duplicates
df = df.drop_duplicates(subset=['Purity', 'Lipid'], keep='last')
#reset index
df.reset_index(drop=True, inplace=True)
#Drop unnamed: 0 column
df.drop(columns=['Unnamed: 0'], inplace=True)


df
```

```python
df_crude = df[df['Purity'] == 'Crude']
df_degummed = df[df['Purity'] == 'Degummed']
df_RBD = df[df['Purity'] == 'RBD']
df_crude
df_degummed
df_RBD
df_combined = pd.DataFrame(columns=['Lipid', 'Crude n-9/n-7', 'Degummed n-9/n-7', 'RBD n-9/n-7'])
df_combined['Lipid'] = df_crude['Lipid']
df_combined['Crude n-9/n-7'] = df_crude['Ratio']
#add the ratio from degummed to the combined dataframe
df_combined['Degummed n-9/n-7'] = df_degummed['Ratio'].values
#add the ratio from RBD to the combined dataframe
df_combined['RBD n-9/n-7'] = df_RBD['Ratio'].values
df_combined['Retention_Time'] = df_crude['Retention_Time'].values
df_combined.to_excel('CLAW_data/CLAW_Canola_OzON_Ratios.xlsx')
df_combined.to_csv('CLAW_data/CLAW_Canola_OzON_Ratios.csv')
df_combined
```

```python
df_OzESI = df_combined.copy()
df_OzESI
```

```python
df_caitlin = pd.read_csv('data/Canola_Caitlin_Purity.csv')

df_caitlin
```

Plotting Crude CLAW vs manual

```python
import matplotlib.pyplot as plt

# Data from OzON ESI
ozesi_lipids = df_OzESI['Lipid']
ozesi_crude_ratio = df_OzESI['Crude n-9/n-7']

# Plotting OzON ESI data with larger dots
plt.plot(ozesi_lipids, ozesi_crude_ratio, 'o-', color='black', label='CLAW', markersize=8)  # Increase markersize here

# Data from Caitlin's dataset
caitlin_lipids = df_caitlin['Lipid']
caitlin_crude_ratio = df_caitlin['Crude n-9/n-7']

# Plotting Caitlin's data with larger dots
plt.plot(caitlin_lipids, caitlin_crude_ratio, 'o-', color='blue', label='Manual', markersize=8)  # Increase markersize here

# Labeling the axes
plt.xlabel('Lipid')
plt.ylabel('Ratio n-9/n-7')

# Adding a title and legend
plt.title('Crude n-9/n-7 Ratios',size=16)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.9))
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
#save the figure as png and pdf
plt.savefig('plots/CLAW/Crude n-9-n-7 Ratios.png', bbox_inches='tight')
plt.savefig('plots/CLAW/Crude n-9-n-7 Ratios.pdf', bbox_inches='tight')

# Display the plot
plt.show()
```
