```python

# Special case: Extract 'Blank' sample
if 'Blank_Blank_Blank_Blank_Blank_N1_Brain_Region' in df.columns:
    samples.append('Blank_Blank')


samples = [col.split('_')[0] + '_' + col.split('_')[1] for col in df.columns if 'Brain_Region' in col]
samples = list(set(samples))  # To get unique samples

brain_regions = ['hippocampus', 'diencephalon', 'cortex', 'cerebellum']

# Create an empty dataframe to store results
result_df = pd.DataFrame(columns=['lipid', 'Class', 'Sample', 'Sum_Intensity'])

for index, row in df.iterrows():
    lipid = row['lipid']
    class_type = row['type']  # Assuming 'type' column represents class
    
    for sample in samples:
        total_intensity = 0  # Initialize total intensity for all brain regions of a sample
        
        for region in brain_regions:
            # Special case for Blank
            if sample == 'Blank_Blank':
                col_name = 'Blank_Blank_Blank_Blank_Blank_N1_Brain_Region'
                total_intensity += row[col_name]
            else:
                # Get columns related to the current sample and brain region
                cols = [col for col in df.columns if sample in col and region in col]
                
                # Sum up the intensities for current lipid and region
                total_intensity += row[cols].sum()
        
        # Append to result_df once per sample (summing across all brain regions)
        result_df = result_df.append({
            'lipid': lipid,
            'Class': class_type,
            'Sample': sample,
            'Brain_Region': 'All',
            'Sum_Intensity': total_intensity
        }, ignore_index=True)

print(result_df)
```

```python
import os

base_filename = "MRM_results/summed_intensity"
extension = ".csv"
counter = 1
output_filename = f"{base_filename}{counter}{extension}"

while os.path.exists(output_filename):
    counter += 1
    output_filename = f"{base_filename}{counter}{extension}"

# Save the dataframe to the new filename
result_df.to_csv(output_filename, index=False)
```

# Subtract the blank

```python
import pandas as pd

# Load your data
df_blanksub = pd.read_csv('MRM_results/summed_intensity4.csv')

# Find the Sum_Intensity for the Blank_Blank sample for each lipid
blank_intensities = df_blanksub[df_blanksub['Sample'] == 'Blank_Blank'].set_index('lipid')['Sum_Intensity']

# Subtract the Blank_Blank Sum_Intensity from each sample's Sum_Intensity
df_blanksub['Adjusted_Intensity'] = df_blanksub.apply(lambda row: row['Sum_Intensity'] - blank_intensities.get(row['lipid'], 0) if row['Sample'] != 'Blank_Blank' else row['Sum_Intensity'], axis=1)

# # Save the adjusted dataframe
# df_blanksub.to_csv('MRM_results/adjusted_intensity.csv', index=False)

df_blanksub.head(None)
```

```python
#drop all rows with the Sample Blank_Blank
df_sub = df_blanksub[df_blanksub.Sample != 'Blank_Blank']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the data
heatmap_data = df_sub.pivot(index='lipid', columns='Sample', values='Adjusted_Intensity')

# Plot the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=.5)
plt.title('Adjusted Intensity Heatmap')
plt.show()
```
