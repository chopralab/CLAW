# Each sample each brain region

```python
brain_regions = ['hippocampus', 'diencephalon', 'cortex', 'cerebellum']

# Extract the sample names (based on the updated naming convention)
samples = set('_'.join(col.split('_')[:2]) for col in df.columns if ('DOD' in col or 'FAD' in col) and ('_F' in col or '_M' in col))

# Iterate through each sample and sum up the counts for the four brain regions
for sample in samples:
    # Find columns matching the sample
    matching_cols = [col for col in df.columns if col.startswith(sample)]
    
    # Sum these columns row-wise and add to a new column
    df[sample + '_summed'] = df[matching_cols].sum(axis=1)

# To check the result
print(df)

```

# Subtract the blank

```python

# Extract the sample names
samples = set('_'.join(col.split('_')[:2]) for col in df.columns if ('DOD' in col or 'FAD' in col) and ('_F' in col or '_M' in col))

# Iterate through each sample and sum up the counts for the four brain regions
for sample in samples:
    # Find columns matching the sample
    matching_cols = [col for col in df.columns if col.startswith(sample)]
    
    # Sum these columns row-wise and add to a new column
    df[sample + '_summed'] = df[matching_cols].sum(axis=1)

# Identify the Blank_Blank column
blank_col = [col for col in df.columns if 'Blank_Blank' in col][0]

# Subtract the Blank_Blank values from each of the summed columns
for col in df.columns:
    if '_summed' in col:
        df[col] -= df[blank_col]

# To check the result
print(df)
```
