# FIND MATHCES BETWEEN OZON AND OZOFF DF

```python
import pandas as pd

# Example DataFrames
# Assuming these DataFrames are already defined:
# OzON_df and OzOFF_df with columns including 'FAC' and 'Peak_Retention_Time'

def find_matches(OzON_df, OzOFF_df, rt_tolerance=0.5):
    # Prepare the output DataFrame with additional columns for match details
    matched_columns = list(OzON_df.columns) + ['OzOFF_Match', 'OFF_RT']
    df_matched = pd.DataFrame(columns=matched_columns)

    # Iterate through each row in the OzON DataFrame
    for index, ozon_row in OzON_df.iterrows():
        on_fac_list = ozon_row['FAC']
        on_rt = ozon_row['Peak_Retention_Time']

        # Check each row in the OzOFF DataFrame for potential FAC matches
        for _, ozoff_row in OzOFF_df.iterrows():
            off_fac_list = ozoff_row['FAC']
            off_rt = ozoff_row['Peak_Retention_Time']

            # Check for any common FAC entries and within the retention time tolerance
            if any(fac in off_fac_list for fac in on_fac_list) and abs(on_rt - off_rt) <= rt_tolerance:
                # Create a new row for the matched DataFrame including additional match details
                match_data = ozon_row.to_dict()
                match_data['OzOFF_Match'] = off_fac_list
                match_data['OFF_RT'] = off_rt
                df_matched = df_matched.append(match_data, ignore_index=True)

    return df_matched

# Usage
df_matched = find_matches(OzON_df, OzOFF_df)
df_matched
```

# Remove OzON with ##:0 

```python
def contains_colon_zero(fac_string):
    return ':0' not in fac_string  # Return True only if ':0' is not found

# Apply the filter to exclude strings containing ':0'
df_filtered = df_matched[df_matched['OzOFF_Match'].apply(contains_colon_zero)]

df_filtered
```

# MATCH FAC TO OzOFF Match

```python
import pandas as pd

# Assuming you have a DataFrame named 'df_filtered' with columns 'FAC' and 'OzOFF_Match'

# Convert the 'OzOFF_Match' column to a list if it's a string
df_filtered['OzOFF_Match'] = df_filtered['OzOFF_Match'].apply(lambda x: [x] if isinstance(x, str) else x)

# Apply the isin() function to check if elements from OzOFF_Match exist in FAC
df_filtered['Match'] = df_filtered.apply(lambda row: any(item in row['FAC'] for item in row['OzOFF_Match']), axis=1)

# Filter the DataFrame to keep only rows with a match
df_filtered_correctly = df_filtered[df_filtered['Match']]

# Drop the 'Match' column since it's no longer needed
df_filtered_correctly = df_filtered_correctly.drop(columns=['Match'])

# Print the filtered DataFrame to verify the results
print("Filtered DataFrame:")
print(df_filtered_correctly)
```

```python
# Columns to keep from the matched DataFrame
selected_columns = [
    'Lipid_ON', 
    'Biology_ON', 
    'Genotype_ON', 
    'Cage_ON', 
    'Mouse_ON', 
    'Peak_Retention_Time_ON', 
    'Peak_Retention_Time_OFF',
    'Peak_Area_ON',
    'Group_Sample_ON',
    'FAC_ON',
    'FAC_OFF'
]

# Subsetting the DataFrame to only include the desired columns
condensed_df = matched_df[selected_columns]

# Renaming the columns to remove the suffixes for cleaner presentation
condensed_df.columns = [
    'Lipid', 
    'Biology', 
    'Genotype', 
    'Cage', 
    'Mouse', 
    'Peak_Retention_Time_ON', 
    'Peak_Retention_Time_OFF',
    'Peak_Area_ON',
    'Group_Sample_ON',
    'FAC_ON',
    'FAC_OFF'
]
selected_biology = 'dienc'
selected_genotype = '5xFAD'  # Assuming you also have a specific genotype to filter
# Filter DataFrame based on the specified 'Biology' and 'Genotype'
condensed_df = condensed_df[(condensed_df['Biology'] == selected_biology) & (condensed_df['Genotype'] == selected_genotype)]

# Display the condensed DataFrame
# Dropping duplicates from the DataFrame and returning a new DataFrame
condensed_df = condensed_df.drop_duplicates()

# Dropping duplicates inplace (modifying the original DataFrame)
condensed_df.drop_duplicates(inplace=True)
condensed_df.head(50)
```
