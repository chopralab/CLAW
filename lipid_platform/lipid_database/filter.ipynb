# OzOFF vs OzON DATABASE

```python
import pandas as pd

def fill_missing_values(df):
    for index, row in df.iterrows():
        if pd.notna(row['Compound Name']) and '...' in row['Compound Name']:
            start, end = row['Compound Name'].split(' / ... / ')
            start_num, start_n = start.split('n-')
            end_num, end_n = end.split('n-')
            
            start_num = int(start_num.split(':')[0])
            end_num = int(end_num.split(':')[0])
            start_n = int(start_n)
            end_n = int(end_n)
            
            missing_values = []
            for i in range(start_num + 1, end_num):
                missing_values.append(f"{i}:1n-{start_n + i - start_num}")
            
            row['Compound Name'] = ' / '.join([start] + missing_values + [end])
    
    return df

# Assuming your Excel file is named 'AMP_Database.xlsx'
df = pd.read_excel('AMP_Database.xlsx')

df = fill_missing_values(df)

print(df)
```
