```python
import pandas as pd
```

```python
df = pd.read_excel('AMP_Database_OzOFF.xlsx')

# Ensure proper column names
df.rename(columns=lambda x: x.strip().replace(' ', '_'), inplace=True)

# Convert necessary columns to numeric, handling non-numeric gracefully
df['Parent_Ion'] = pd.to_numeric(df['Parent_Ion'], errors='coerce')
df['Product_Ion'] = pd.to_numeric(df['Product_Ion'], errors='coerce')
# Fill NaN values in 'ISTD?' with a default value and convert to string
df['ISTD?'] = df['ISTD?'].fillna('None').astype(str)
df.to_parquet('Database_OzOFF.parquet', index=False)
df
```

```python
import pandas as pd


# Rename 'Compound Name' to 'Lipid'
df.rename(columns={'Compound_Name': 'Lipid'}, inplace=True)

# Update 'Lipid' column to wrap the existing values with 'FA()'
df['Lipid'] = df['Lipid'].apply(lambda x: f'FA({x})')


# Create 'Chain_length' column by extracting the chain length number from the 'Lipid' column
df['Chain_length'] = df['Lipid'].str.extract(r'FA\((\d+):')

df.drop(columns=['ISTD?'], inplace=True)


df
```
