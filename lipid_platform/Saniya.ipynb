```python
import os
import pandas as pd
import re

# Function to process the Sample column in a DataFrame
def process_sample_column(df):
    # Define the regex patterns for Sample_Name and Sup_Pel
    sample_name_pattern = re.compile(r'(SV\d+|DMSO|MED|Blank)', re.IGNORECASE)
    sup_pel_pattern = re.compile(r'_(s\d+|p\d+)')
    
    # Initialize new columns
    df['Sample_Name'] = df['Sample'].apply(lambda x: sample_name_pattern.search(x).group(0) if sample_name_pattern.search(x) else None)
    df['Sup_Pel'] = df['Sample'].apply(lambda x: sup_pel_pattern.search(x).group(1) if sup_pel_pattern.search(x) else None)
    
    return df

# Function to iterate through parquet files, load them, and process the Sample column
def combine_and_process_parquet_files(directory_path):
    # Initialize an empty list to hold DataFrames
    dataframes = []
    file_name_map = {}  # Dictionary to map Sample_Name to file names
    
    # Iterate through all files in the directory
    for filename in os.listdir(directory_path):
        # Only process parquet files
        if filename.endswith('.parquet'):
            file_path = os.path.join(directory_path, filename)
            # Load the parquet file into a DataFrame
            df = pd.read_parquet(file_path)
            # Check if 'Sample' column exists
            if 'Sample' in df.columns:
                # Process the Sample column in the DataFrame
                df = process_sample_column(df)
                # Map the Sample_Name to the file name
                unique_sample_names = df['Sample_Name'].unique()
                for sample_name in unique_sample_names:
                    file_name_map[sample_name] = filename
                dataframes.append(df)
            else:
                # Print the file name if the Sample column is missing
                print(f"File '{filename}' does not have a 'Sample' column.")
    
    # Concatenate all DataFrames into one
    if dataframes:
        combined_df = pd.concat(dataframes, ignore_index=True)
        return combined_df, file_name_map
    else:
        print("No valid DataFrames to combine.")
        return pd.DataFrame(), {}  # Return an empty DataFrame if no files were processed

# Function to perform blank subtraction based on user-specified blank sample
def blank_subtract(combined_df, blank_sample_name, file_name_map):
    # Print the file being used for the blank sample
    if blank_sample_name in file_name_map:
        print(f"File being used for the blank sample: {file_name_map[blank_sample_name]}")
    else:
        print(f"Blank sample '{blank_sample_name}' not found in file map.")
    
    # Filter the DataFrame to get the blank sample data
    blank_df = combined_df[combined_df['Sample_Name'] == blank_sample_name]
    
    # Create a dictionary with Lipid as the key and the corresponding OzESI_Intensity of the blank as the value
    blank_intensities = blank_df.set_index('Lipid')['OzESI_Intensity'].to_dict()
    
    # Initialize a new DataFrame to hold the result after blank subtraction
    combined_df_blanksubtract = combined_df.copy()

    # Perform blank subtraction for all rows where Sample_Name is not the blank
    combined_df_blanksubtract['Intensity_Subtracted'] = combined_df_blanksubtract.apply(
        lambda row: row['OzESI_Intensity'] - blank_intensities.get(row['Lipid'], 0) 
        if row['Sample_Name'] != blank_sample_name else row['OzESI_Intensity'], axis=1)
    
    return combined_df_blanksubtract

# Example usage
directory_path = 'Projects/saniya/match/OFF/'
combined_df, file_name_map = combine_and_process_parquet_files(directory_path)

# Call the blank_subtract function
blank_sample_name = 'Blank'  # User-specified blank sample name
combined_df_blanksubtract = blank_subtract(combined_df, blank_sample_name, file_name_map)
```

```python
import os
import zipfile
import pandas as pd

def convert_parquet_to_csv(parquet_file, csv_file):
    """
    Converts a Parquet file to a CSV file.
    
    Parameters:
    - parquet_file (str): The path to the Parquet file.
    - csv_file (str): The output path for the CSV file.
    """
    # Read the Parquet file into a DataFrame
    df = pd.read_parquet(parquet_file)
    # Save the DataFrame as a CSV file
    df.to_csv(csv_file, index=False)

def compress_parquet_files_to_csv(directory_path, output_zip):
    """
    Converts all .parquet files in the given directory to CSV and compresses them into a zip file.
    
    Parameters:
    - directory_path (str): The path to the directory containing Parquet files.
    - output_zip (str): The output zip file path (including the filename).
    """
    # Create a ZipFile object for writing compressed files
    with zipfile.ZipFile(output_zip, 'w') as zipf:
        # Iterate over all files in the directory
        for root, dirs, files in os.walk(directory_path):
            for file in files:
                if file.endswith('.parquet'):
                    # Get the full file path for the Parquet file
                    parquet_file = os.path.join(root, file)
                    
                    # Create a temporary CSV file path
                    csv_file = os.path.splitext(parquet_file)[0] + '.csv'
                    
                    # Convert Parquet to CSV
                    convert_parquet_to_csv(parquet_file, csv_file)
                    
                    # Add the CSV file to the zip archive
                    zipf.write(csv_file, os.path.basename(csv_file))
                    
                    # Remove the temporary CSV file after adding it to the zip archive
                    os.remove(csv_file)
                    
                    print(f"Converted and compressed: {file} to {os.path.basename(csv_file)}")
    
    print(f"All Parquet files in {directory_path} have been converted to CSV and compressed into {output_zip}.")

# Example usage
directory_path = 'Projects/saniya/match/OFF/'  # Specify the directory containing the Parquet files
output_zip = 'Projects/saniya/saniya_FA_csv.zip'  # Specify the output compressed zip file

compress_parquet_files_to_csv(directory_path, output_zip)
```

```python
import os
import zipfile
import pandas as pd

def convert_parquet_to_csv(parquet_file, csv_file):
    """
    Converts a Parquet file to a CSV file.
    
    Parameters:
    - parquet_file (str): The path to the Parquet file.
    - csv_file (str): The output path for the CSV file.
    """
    # Read the Parquet file into a DataFrame
    df = pd.read_parquet(parquet_file)
    # Save the DataFrame as a CSV file
    df.to_csv(csv_file, index=False)

def compress_parquet_files_to_csv(directory_path, output_zip):
    """
    Converts all .parquet files in the given directory to CSV and compresses them into a zip file.
    
    Parameters:
    - directory_path (str): The path to the directory containing Parquet files.
    - output_zip (str): The output zip file path (including the filename).
    """
    # Initialize a counter for the number of files added to the zip
    file_count = 0
    
    # Create a ZipFile object for writing compressed files
    with zipfile.ZipFile(output_zip, 'w') as zipf:
        # Iterate over all files in the directory
        for root, dirs, files in os.walk(directory_path):
            for file in files:
                if file.endswith('.parquet'):
                    # Get the full file path for the Parquet file
                    parquet_file = os.path.join(root, file)
                    
                    # Create a temporary CSV file path
                    csv_file = os.path.splitext(parquet_file)[0] + '.csv'
                    
                    # Convert Parquet to CSV
                    convert_parquet_to_csv(parquet_file, csv_file)
                    
                    # Add the CSV file to the zip archive
                    zipf.write(csv_file, os.path.basename(csv_file))
                    
                    # Remove the temporary CSV file after adding it to the zip archive
                    os.remove(csv_file)
                    
                    # Increment the counter
                    file_count += 1
                    
                    print(f"Converted and compressed: {file} to {os.path.basename(csv_file)}")
    
    # Print the total number of files added to the zip
    print(f"All Parquet files in {directory_path} have been converted to CSV and compressed into {output_zip}.")
    print(f"Total unique files added to the zip: {file_count}")

# Example usage
directory_path = 'Projects/saniya/match/OFF/'  # Specify the directory containing the Parquet files
output_zip = 'Projects/saniya/saniya_FA_csv.zip'  # Specify the output compressed zip file

compress_parquet_files_to_csv(directory_path, output_zip)
```

```python
import os

def count_files_in_directory(directory_path):
    """
    Counts and prints the number of files in the given directory (non-recursive)
    and prints the name of each file.
    
    Parameters:
    - directory_path (str): The path to the directory to count files in.
    """
    # List all files in the directory
    files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
    
    # Print each file name
    print(f"Files in {directory_path}:")
    for file in files:
        print(file)
    
    # Print the total number of files
    file_count = len(files)
    print(f"\nTotal number of files in {directory_path}: {file_count}")

# Example usage
directory_path = 'Projects/saniya/match/OFF/'  # Specify the directory path
count_files_in_directory(directory_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import numpy as np

# Directory containing the parquet files
directory_path = 'Projects/saniya/match/OFF/'
output_file = os.path.join(directory_path, 'saniya_FA_heatmap.png')

# Define the blank file name
blank_file = 'df_match_3_FA_AMP_Blank_N1_9_6_24_summed.parquet'
blank_file_path = os.path.join(directory_path, blank_file)

# Read the blank parquet file into a DataFrame
blank_df = pd.read_parquet(blank_file_path)

# Initialize an empty DataFrame to store the aggregated data for heatmap
heatmap_data = pd.DataFrame()

# Find all parquet files in the directory
parquet_files = glob.glob(os.path.join(directory_path, '*.parquet'))

# Remove the blank file from the list of files to process
parquet_files.remove(blank_file_path)

# Perform log transformation on blank intensity
blank_df['OzESI_Intensity_Log2'] = np.log10(blank_df['OzESI_Intensity'].replace(0, np.nan))

# Create a dictionary for blank intensities keyed by lipid for fast lookup
blank_intensities = blank_df.set_index('Lipid')['OzESI_Intensity_Log2'].to_dict()

# Iterate through each parquet file (excluding the blank)
for file in parquet_files:
    # Read the parquet file into a DataFrame
    df = pd.read_parquet(file)
    
    # Print the columns of the current DataFrame for debugging
    print(f"File: {file}")
    print(f"Columns: {df.columns}")
    
    # Check if 'Sample' column exists
    if 'Sample' in df.columns:
        # Extract Sample name (assuming it's the same for the whole file)
        sample_name = df['Sample'].iloc[0]
    else:
        print(f"'Sample' column not found in file: {file}")
        continue  # Skip this file and move to the next one
    
    # Apply log transformation to the sample intensity
    df['OzESI_Intensity_Log2'] = np.log10(df['OzESI_Intensity'].replace(0, np.nan))  # Replace zeros with NaN to avoid log(0)
    
    # Perform blank subtraction
    df['Intensity_Subtracted'] = df.apply(lambda row: row['OzESI_Intensity_Log2'] - blank_intensities.get(row['Lipid'], np.nan), axis=1)
    
    # Add the data to heatmap_data DataFrame
    for lipid, intensity in zip(df['Lipid'], df['Intensity_Subtracted']):
        heatmap_data.loc[lipid, sample_name] = intensity

# Fill any missing values with zeros (or you can keep NaN if you prefer)
heatmap_data.fillna(0, inplace=True)

# Plot the heatmap with red and blue color map
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="seismic", cbar_kws={'label': 'Log10(Intensity - Blank)'}, annot=False)

# Label the axes
plt.xlabel('Sample', fontsize=12)
plt.ylabel('Lipid', fontsize=12)
plt.title('Lipid Intensity Heatmap (Log10) with Blank Subtraction', fontsize=14)

# Show the heatmap
plt.tight_layout()
plt.savefig(output_file, dpi=300)  # Save with high resolution (300 dpi)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np

# Directory containing the parquet files and where the heatmap will be saved
directory_path = 'Projects/saniya/match/OFF/'
output_file = os.path.join(directory_path, 'saniya_FA_heatmap_grouped.png')

# Function to group p* together and s* together in Sup_Pel
def group_sup_pel(sup_pel, sample_name):
    if sample_name == "Blank":
        return None  # Skip processing for Blank samples
    if pd.isna(sup_pel):
        return None  # Skip if Sup_Pel is None or NaN
    if sup_pel.startswith('p'):
        return 'p_group'
    elif sup_pel.startswith('s'):
        return 's_group'
    return sup_pel  # In case there are any other categories, keep them as is

# Assume 'combined_df' is already defined and includes 'Sample_Name' and 'Sup_Pel' columns
# For this example, let's use combined_df directly in place of iterating through files

# Perform log transformation on OzESI_Intensity
combined_df['OzESI_Intensity_Log2'] = np.log10(combined_df['OzESI_Intensity'].replace(0, np.nan))

# Create a dictionary for blank intensities keyed by lipid for fast lookup
# Assuming `blank_df` has already been created as in your original code
blank_intensities = blank_df.set_index('Lipid')['OzESI_Intensity_Log2'].to_dict()

# Apply grouping to 'Sup_Pel' and check for 'Sample_Name' being "Blank"
combined_df['Sup_Pel_Grouped'] = combined_df.apply(
    lambda row: group_sup_pel(row['Sup_Pel'], row['Sample_Name']), axis=1
)

# Perform blank subtraction
combined_df['Intensity_Subtracted'] = combined_df.apply(
    lambda row: row['OzESI_Intensity_Log2'] - blank_intensities.get(row['Lipid'], np.nan), axis=1
)

# Initialize an empty DataFrame to store the aggregated data for heatmap
heatmap_data = pd.DataFrame()

# Group the DataFrame by 'Sample_Name', 'Sup_Pel_Grouped', and 'Lipid'
grouped_df = combined_df.groupby(['Sample_Name', 'Sup_Pel_Grouped', 'Lipid'])

# Calculate the average intensity for each group
average_intensity_df = grouped_df['Intensity_Subtracted'].mean().reset_index()

# Add the data to heatmap_data DataFrame
for _, row in average_intensity_df.iterrows():
    lipid = row['Lipid']
    sample_group = f"{row['Sample_Name']}_{row['Sup_Pel_Grouped']}"
    intensity = row['Intensity_Subtracted']
    heatmap_data.loc[lipid, sample_group] = intensity

# Fill any missing values with zeros (or you can keep NaN if you prefer)
heatmap_data.fillna(0, inplace=True)

# Plot the heatmap with a red-blue color map
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="seismic", cbar_kws={'label': 'Log10(Intensity - Blank)'}, annot=False)

# Label the axes
plt.xlabel('Sample Group', fontsize=12)
plt.ylabel('Lipid', fontsize=12)
plt.title('Lipid Intensity Heatmap (Log10) with Blank Subtraction and Grouped Samples (Averaged)', fontsize=14)

# Show and save the heatmap
plt.tight_layout()
plt.savefig(output_file, dpi=300)  # Save with high resolution (300 dpi)
plt.show()
```
