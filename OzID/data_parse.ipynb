```python
#TG df
df_TG = pd.DataFrame()

#create columns for df
#TG = tags | m/z = mass to charge ratio
df_TG['TG_40'] = ['TG_40_{}'.format(i) for i in range(1,12)]
df_TG['TG_40(m/z)'] = [(712.6 - (2 * i)) for i in range(1,12)]
df_TG['TG_41'] = ['TG_41_{}'.format(i) for i in range(1,12)]
df_TG['TG_41(m/z)'] = [(712.6 - (2 * i)) for i in range(1,12)]
df_TG
```

```python
#Oz df
df_TG_Oz = pd.DataFrame()
df_TG_Oz['n-?'] = [('n-{}'.format(i)) for i in range(1,12)]
df_TG_Oz['TG_40_1(m/z)'] = [(726.6 - (14 * i)) for i in range(1,12)]

for i in range(2,10):
    count = 726.6
    count -= 2*i
    df_TG_Oz['TG_40_{}(m/z)'.format(i)] = [(count -(14 * i)) for i in range(1,12)]
df_TG_Oz
```
